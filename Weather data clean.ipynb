### Imports and loading data from csv

```python
import pandas as pd
import numpy as np
import math
```

```python
raw_weather = pd.read_csv("GwangyangWeather.csv")
display(raw_weather)
```

```python
date_weather = raw_weather.copy()
display(date_weather)
```

### Creating some derieved columns

```python
def getYearfromColumn(weatherDateTime):
    results = weatherDateTime.split("-")
    return results[0]

def getMonthfromColumn(weatherDateTime):
    results = weatherDateTime.split("-")
    return results[1]

def getDayfromColumn(weatherDateTime):
    results = weatherDateTime.split("-")
    return results[2]

def getSunriseHour(sunrise):
    time = sunrise.split("T")
    results = time[1].split(":")
    return results[0]

def getSunsetHour(sunset):
    time = sunset.split("T")
    results = time[1].split(":")
    return results[0]
```

```python
date_weather["Year"] = date_weather["datetime"].apply(getYearfromColumn)
date_weather["Month"] = date_weather["datetime"].apply(getMonthfromColumn)
date_weather["Day"] = date_weather["datetime"].apply(getDayfromColumn)

date_weather["SunriseHour"] = date_weather["sunrise"].apply(getSunriseHour)
date_weather["SunsetHour"] = date_weather["sunset"].apply(getSunsetHour)

display(date_weather)
```

```python
print(date_weather.columns)
print(len(date_weather.columns))
```

### Filtering out some columns

```python
display(date_weather.loc[:,['severerisk', 'sunrise','sunset','moonphase','conditions']].head())
```

```python
weather_columns = date_weather.loc[:, ['tempmax','tempmin', 'temp', 
                                   'feelslikemax', 'feelslikemin', 'feelslike', 'dew', 'humidity',
                                  'precip', 'precipprob', 'precipcover', 'preciptype',
                                  'snowdepth', 'windgust', 'windspeed', 'winddir', 'sealevelpressure',
                                  'cloudcover', 'visibility', 'solarradiation', 'solarenergy', 'uvindex',
                                  'conditions',
                                  'Year', 'Month',
                                  'Day', 'SunriseHour', 'SunsetHour']]

display(weather_columns)
```

### Filling in 0 values and NaN values

```python
def FillZeroInNan(val):
    if math.isnan(val):
        return 0
    else:
        return val

weather_columns["snowdepth"] = weather_columns["snowdepth"].apply(FillZeroInNan)
weather_columns["windgust"] = weather_columns["windgust"].apply(FillZeroInNan)

#display(weather_columns["windgust"])
```

```python
def FillClearInPrecipType(val):
    if val is np.nan:
        return 'noprecip'
    else:
        return val
    
weather_columns["preciptype"] = weather_columns["preciptype"].apply(FillClearInPrecipType)
```
